Here we will 1) load original datasets, 2) remove unused columns, 3) de-duplicate rows, and 4) join datasets, not necessarily in this order

Assumptions: 
- Claim ID (cur_clm_uniq_id) represents one claim, which may or may not have more than one code (code could be HCPCS/CPT, diagnosis, procedure...)
- There is a one-to-many relationship between patient IDs (bene_mbi_id) and claim IDs (cur_clm_uniq_id), i.e. each claim is unique to one patient, but one patient can have more than one claim

### 0.3 Combine all datasets

### 0.3.1. Join datasets with common records

```python
# Join Claims Header and Claims Revenue Center datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    parta_claims_header_df,
    parta_claims_revenue_center_detail_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

```python
# Join Medicare and Diagnosis datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    medicare_df,
    parta_diagnosis_code_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

```python
# Since Claims Header dataset has some principal diagnosis codes and the Diagnosis dataset 
#  supplements them with additional codes wherever possible, 
#  coalesce them with preference to the principal code from Claim Header dataset
# Once the diagnosis codes are combined into one column, remove the older columns and any duplicates
medicare_df['diagnosis_code'] = medicare_df[['prncpl_dgns_cd', 'clm_dgns_cd']].bfill(axis=1).iloc[:, 0]
medicare_df = medicare_df.drop(['prncpl_dgns_cd', 'clm_dgns_cd'], axis=1).drop_duplicates()
medicare_df.sort_values(by='claim_id').head(20)
```

```python
# Since Claims Header dataset records the amount Medicare paid for the claims 
#  and Claims Revenue Center dataset records the amount Medicare reimbursed the provider,
#  assume they were separate charges and add them to get the total cost for claim (for particular code)
# Once the costs are combined into one column, remove the older columns and any duplicates
medicare_df['claim_cost'] = medicare_df['clm_pmt_amt']+medicare_df['clm_line_cvrd_pd_amt']
medicare_df = medicare_df.drop(['clm_pmt_amt', 'clm_line_cvrd_pd_amt'], axis=1).drop_duplicates()
medicare_df.sort_values(by='claim_id').head(20)
```

```python
def create_claims_comparison_plot(
    cvm_data: pd.Series,
    blood_data: pd.Series,
    chemistry_data: pd.Series,
    pathology_data: pd.Series,
    medicare_data: pd.Series,
    figsize: tuple = (10, 6)
) -> None:

    # Create figure
    plt.figure(figsize=figsize)
    
    # Plot horizontal bars for each claim type
    bars = {
        'CVM Claims': (cvm_data, 1.0),
        'Blood Claims': (blood_data, 1.0),
        'Chemistry Claims':(chemistry_data,1.0),
        'Pathology Claims':(pathology_data,1.0),
        'Other Claims': (medicare_data, 0.5)
    }
    
    for label, (data, alpha) in bars.items():
        plt.barh(y=data.index, width=data.values, label=label, alpha=alpha)
    
    # Add percentage labels for CVM claims
    for i, value in enumerate(cvm_data.values):
        total_claims = cvm_data.values[i] + medicare_data.values[i]
        percentage = (value / total_claims) * 100
        plt.text(value, i, f' {percentage:.1f}%', va='center')
    
    # Customize plot
    plt.xlabel('Number of Claims')
    plt.ylabel('Year')
    plt.title('CVM claims as a percentage of total claims over the years')
    plt.legend()
    
    plt.show()
```

## Q2. Evaluate the HCP behavior in context of claim volume from 2016-2018. How many HCPs are submitting 1 CVM claim; how many HCPs are associated with more than 10 claims, etc.? Once you perform this analysis, explain how this trend can influence the sales force deployment. That is, how would you segment the HCPs and how would you allocate In-Person (sales force) vs Non-Personal Promotions (NPP, i.e. Emails, Social Media, Digital etc.) efforts?

```python
def create_plot_axis(figsize: tuple = (10, 6)) -> plt.Axes:
    fig, ax = plt.subplots(figsize=figsize)
    return ax

ax = create_plot_axis()

hcp_claim_counts=cvm_medicare.groupby('npi_id').size()
hcp_category_counts = pd.cut(hcp_claim_counts, bins=[0, 1, 4, 9, float('inf')], 
                             labels=['1 Claim', '2-4 Claims', '5-9 Claims', '10+ Claims']).value_counts().sort_index()
ax.barh(y=range(len(hcp_category_counts)), 
        width=hcp_category_counts.values, 
        color='steelblue')
ax.set_yticks(range(len(hcp_category_counts)))
ax.set_yticklabels(hcp_category_counts.index)

for i, count in enumerate(hcp_category_counts.values):
    ax.text(count + 2, i, f'{count}', ha='left', va='center', fontsize=12)

ax.set_ylabel('Claim Volume', fontsize=12)
ax.set_xlabel('HCPs Count', fontsize=12)
ax.set_title('HCP behavior in context of claim volume from 2016-2018', fontsize=14)

plt.show()
```

### Business Insights:

As we can see in the plot above, HCPs that have only 1 claim in HCP are the highest in number. The least are the ones with more than 10+ claims. There is a clear decline in the number of repeat claims.

A quick review suggests that the majority of HCPs (over 6,200) have only one experience with CVM claims while a small number of 150 HCPs (5-9 or 10 claims) are high-volume prescribers. This is an example of a classic "long-tail" scenario, where a small number of HCPs contribute a large majority to total claim activity.

1. High-Volume HCPs (5 Claims)
The highest prescribing HCPs (up to 10 claims) are often specialists or, in primary care, HCPs that manage a large panel of patients.
Selling Approach:In-Person Focused: This group should be visited on a relatively frequent basis, have specialized medical education events, and could likely engage in deeper scientific discussions. This group enjoys the data displayed to them with advanced clinical data, and with some, with real-world evidence, in addition to deeper relationships with the rep.
KOL Development: Identify prospective "Key Opinion Leaders" who may share best practices and/or stories of success with their peer group.

3. Moderate-Volume HCPs (2–4 Claims)
They may have some experience with CVM examples but are not yet high-volume prescribers.
Selling Approach:
Hybrid Approach - An occasional in-person visit, to address concerns (e.g., concerns on reimbursement). Near the earlier point of sale, we can also emphasize specific examples on how a CVM has benefited a specific patient
Follow-Up Via Targeted Digital Campaign - This wouldn't be in-person follow-up but e-detailing, training for online modules, or a newsletter via email, to reinforce the information presented in the meeting. 

3. HCPs with Few Claims (1 Claim)
This is by far the largest segment (6,246 HCPs); they wrote just one CVM claim in three years, and are likely dabblers or have a limited relevant patient population. Sales Approach: Non-Personal Promotion (NPP): Given the size of the segment, broad-reaching digital tactics including emails, social media campaigns and short educational videos are more tractable and cost effective than mass field visits. Onboarding and Awareness: Distill your clinical message to easily digestible pieces, include a brief summary of coverage processes, and have step-by-step guidance for their prescribing. The goal will be to either convert them back to prescribing more regularly or find ways for them to prescribe more frequently with easy to consume content.

Segmentation & Resource Allocation Strategy:
Provide more field rep coverage to high-tier prescribers (5+ claims) as personal engagement and deeper interaction is more likely to drive sizable prescription growth or have them advocate for your brand. As for the 2-4 claim HCPs, take a hybrid approach of well-timed in-person reps, to establish rapport, combined with a digitally convenient ongoing connection. This provides a good cost-effective balance, with a chance of moving them up the usage curve. For the one-claim HCP, mainly rely on NPP methods – mainly automatic email drip campaigns, short explainer videos, and tele-detailing while trying to re-engage them with consideration of common pain points (coverage or formulary misconceptions). 



## Q3. Evaluate the Patient Age demographics in the context of claim volume from 2016-2018. Bucket the patients into groups based on their age and explain the trends. How would you position the Marketing Budgets and the Promotions with respect to the changing landscape of the CVM claims and the respective patient segments? 

```python
def create_age_distribution_plot(data: pd.DataFrame, 
                               age_column: str,
                               ax: plt.Axes,
                               age_bins: list[int] = [18, 40, 60, 80, 101],
                               age_labels: list[str] = ['18-39', '40-59', '60-79', '80-100']) -> None:
    # Filter valid age records and create age groups
    valid_claims = data[data[age_column] > 0].copy()
    claims_count = pd.cut(valid_claims[age_column], 
                         bins=age_bins, 
                         labels=age_labels).value_counts().sort_index()
    
    # Creating the plot
    ax.barh(y=range(len(claims_count)), 
            width=claims_count.values, 
            color='steelblue')
    
    ax.set_yticks(range(len(claims_count)))
    ax.set_yticklabels(claims_count.index)
    
    # Add value labels to bars
    for i, count in enumerate(claims_count.values):
        ax.text(count + 2, i, f'{count}', ha='left', va='center', fontsize=12)
    
    ax.set_ylabel('Claim Volume', fontsize=12)
    ax.set_xlabel('HCPs Count', fontsize=12)
    ax.set_title('HCP behavior in context of claim volume from 2016-2018', fontsize=14)
fig, ax = plt.subplots(figsize=(10, 6))
create_age_distribution_plot(cvm_medicare, 'patient_age', ax)
plt.show()
```

### Business Insights:

When we categorize patients by age—say, 18–39, 40–59, 60–79, and 80–100—a distinct trend is apparent: claims sharply increase as age rises until a notable peak in the 60–79 age category, and are still notable among people aged 80 and older. This makes sense given the expected increase in chronic conditions, with cardiovascular problems at the forefront, in later life. The younger age groups (especially 18–39) tend to generate fewer claims, which could be related to a lower clinical risk or waiting to seek care until absolutely necessary.

18-39 (Emerging Risk): 
The claims volume is low for this group; however, many individuals in this group may be at early stages of cardiovascular risk related to lifestyle (e.g., sedentary, poor diet). This is an opportunity for intervention and brand building with younger populations. 

40-59 (Developing Health Awareness): 
Individuals approach middle age and become more aware of health concerns, which is associated with increased rates of screening and diagnostic claims. This cohort is more receptive to messaging related to health (e.g., awareness campaigns, lifestyle management). 

60-79 (Peak Claim Volume): 
This is the primary cohort for CVM claims. Comorbidities and chronic condition management (e.g., hypertension, hyperlipidemia, etc.) drive increased frequency and complexity of claims. 

80-100 (Complex Chronic Care): 
While slightly smaller in absolute numbers than the cohort 60-79, patients in the cohort are likely to require more integrated and comprehensive care. Top obstacles may include medication adherence, polypharmacy, and care coordination.

### Positioning Marketing Budgets and Promotions

Preventive & Early Intervention Outreach (18–39, 40–59)Digital-First Strategy: 
This demographic prefers social media, electronic apps or telehealth strategies. Marketing campaigns discussing lifestyle tips, heart-healthy habits, and well visits can help shape a lifelong preference for your brand. Educational Messaging Utilize short infographic, blog-posts, or webinars providing clarity for early indicators of CVM conditions. Getting younger patients to use prevention strategies early on can avoid costly interventions later. 

Core Chronic Disease Management (60–79)Integrated Care Programs: 
Invest a portion of the budget to support chronic disease management, i.e. medication adherence programs, remote monitoring apps or devices, or workspace support teams Multichannel Patient Engagement: Traditional marketing (ie, mailers, television) can apply to older generations, yet older adults are quite tech shy. Incorporating tools such as patient portal and telemedicine can still support your ongoing follow-up. 

High-Touch Support for Older Age (80–100)Tailor Services & Partnerships: 
At older ages, mobility and care complexities are important. Utilization of partnerships with home-care agencies, assisted living facilities, and community health organizations are important to access patients. Concise Messaging: Your message should utilize straightforward calls to action and easy to use learning materials, to help patients and caregivers navigate coverage and care options.

Flexibility & Cross-Generational Alignment:
Be prepared for demographic overlap—patients in their 40s may be assisting aging parents in their 70s or 80s. Tailored but cohesive messaging across channels helps build trust and brand consistency within households.
